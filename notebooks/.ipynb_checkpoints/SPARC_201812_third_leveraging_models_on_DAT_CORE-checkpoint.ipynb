### Concepts

- Model : structured way to define and describe a piece of information.  
    _Think a table in a SQL database or in a EXCEL file._

- Property :  the building block used to describe each instances of a model.  
    _Think a field in the SQL table or a column in a EXCEL file._

- Record : one piece of information that follows the definition given in the model.  
    _Think one row in a SQL table or EXCEL file_

- Relationship : a way to create a named link between two records.  
    _Example: subject 10 __has_kynematic_recording__ in file xxxxx.yyy_

We want to load the json files for the three subjects:
- subject_1/subject1.json
- subject_2/subject2.json
- subject_3/subject3.json

We find the full path for the files in our local copy of the dataset

```python
# initialize list of json files
subjectJsonFileNames = []
# traverse all the local folder structure
for root, dirs, filenames in os.walk(DATASET_PATH):
    # for each folder, extract only the json files according to the pattern subjecX.json
    # add the root path and insert the complete file path in the list
    subjectJsonFileNames += [os.path.join(root,filename) for filename in filenames if filename.startswith('subject') and filename.endswith('.json')]
subjectJsonFileNames
```

Now we load the three json files

```python
subjectJsonContent = []
for filename in subjectJsonFileNames:
    with open(filename,'r') as fh:
        subjectJsonContent.append(json.load(fh))
```

Let's check if the properties (aka fields) are consistent across the three subjects.  
Find all the fields, their type and if they are present in all the subjects

```python
schemaSubject = {}
# loop on every subject info
for subject in subjectJsonContent:
    # check every key
    for key in subject.keys():
        # check if we need to initialize the key
        if key not in schemaSubject.keys():
            schemaSubject[key] = []
        # insert current key and data type
        schemaSubject[key].append(type(subject[key]))
```

```python
schemaSubjectStats = {key : { 'types' : list(set(value)), 'presence' : len(value)} for key,value in schemaSubject.items()}
```

Here is the code to programmatically create the model

```python
dcDataset.update()
mSubject = dcDataset.create_model(DC_MODEL_SUBJECT)
mSubject
```

Now that we have the model, we add all the fields that we decided to include.  
We are excluding the _emgs channels_ and _raw files_ fields. We will deal with them later.

```python
Image(filename=DC_IMAGE_MODEL_2)
```

Here is the code to add the properties to the model (aka fields to our schema)

```python
# add a mandatory field called Name that is designated as title
mSubject.add_property('Name',str,title=True)
# loop on all the keys that are present in the info
for key, item in schemaSubjectStats.items():
    # skip the more complicated fields
    if key in ['emgs_channels', 'raw_files']:
        continue
    mSubject.add_property(key,item['types'][0])
```

...and we verify that the model is complete

```python
for key,item in mSubject.schema.items():
    print(key)
```

We defined the model and its properties, 
now we can upload the 3 subjects data as individual records

```python
Image(filename=DC_IMAGE_RECORDS_1)
```

Here is the code to upload the three records. One for each subject.

```python
for subject in subjectJsonContent:
    # copy the dictionary
    record = copy.deepcopy(subject)
    # remove fields that are not to be uploaded
    record.pop('emgs_channels')
    record.pop('raw_files')
    # add value for field Name
    record['Name'] = '-'.join([str(record['subject_id']),record['subject_name']])
    # create record on balckfynn
    bfRecord = mSubject.create_record(record)
    # save bf id locally
    subject['bf_id'] = bfRecord.id
mSubject.update()
```

We want to verify that the three records have been created...

```python
for record in mSubject.get_all():
    print(record.values['Name'])
```

Let's follow for the EMGs the same process that we did for Subjects

```python
Image(filename=DC_IMAGE_RECORDS_2)
```

First: we build the schema for the emgs

```python
schemaEmgs = {}
# loop on every subject info
for subject in subjectJsonContent:
    for emg in subject['emgs_channels']:
        # check every key
        for key in emg.keys():
            # check if we need to initialize the key
            if key not in schemaEmgs.keys():
                schemaEmgs[key] = []
                # insert current key and data type
            schemaEmgs[key].append(type(emg[key]))
schemaEmgsStats = {key : { 'types' : list(set(value)), 'presence' : len(value)} for key,value in schemaEmgs.items()}
schemaReport(schemaEmgsStats)
```

Second: we create the model

```python
dcDataset.update()
mEmg = dcDataset.create_model(DC_MODEL_EMG)
mEmg
```

Third: add properties to the model

```python
# add a mandatory field called Name that is designated as title
mEmg.add_property('Name',str,title=True)
# loop on all the keys that are present in the info
for key, item in schemaEmgsStats.items():
    mEmg.add_property(key,item['types'][0])
```

Fourth: we upload all the EMG records

```python
subjectVsEmg = {}
for subject in subjectJsonContent:
    subjectVsEmg[subject['bf_id']] = []
    for emg in subject['emgs_channels']:
        # copy the dictionary
        record = copy.deepcopy(emg)
        # add value for field Name
        record['Name'] = '-'.join([str(subject['subject_id']),subject['subject_name'],'channel',str(emg['channel'])])
        # create record on balckfynn
        bfRecord = mEmg.create_record(record)
        subjectVsEmg[subject['bf_id']].append(bfRecord.id)
        mEmg.update()
```

Fifth: we check

```python
mEmg.update()
emgRecords = mEmg.get_all()
len(emgRecords)
```

Now, we have the following individual entities / pieces of information on DAT CORE:
- subject records
- EMG records
- files

Leveraging the relationship functionality in DAT CORE,
we can:
- relate each EMG record with the matching subject record
- relate each file with the matching subject record

Let's create or retrieve the relationships between subjects and the related emgs

```python
if DC_SUBJECT_TO_EMG_NAME in dcDataset.relationships().keys():
    emgRelationship = dcDataset.get_relationship(DC_SUBJECT_TO_EMG_NAME)
else:
    emgRelationship = dcDataset.create_relationship_type(DC_SUBJECT_TO_EMG_NAME,DC_SUBJECT_TO_EMG_DESCRIPTION)
emgRelationship
```

Use the relationships just created to create a relationship between subject and emg channel records.  
Retrieve models for Subject and EMG

```python
mSubject = dcDataset.get_model(DC_MODEL_SUBJECT)
mEmg = dcDataset.get_model(DC_MODEL_EMG)
```

Loop on subject,emg and create relationship

```python
# loop on all the subjects
for sKey,emgs in subjectVsEmg.items():
    # retrieve subject record
    rSubject = mSubject.get(sKey)
    # loop on all subject emgs
    for eKey in emgs:
        # retrieve emg record
        rEmg = mEmg.get(eKey)
        # create relationship between the 2 records
        rSubject.relate_to(rEmg,emgRelationship)
        # sync with DAT CORE
        rEmg.update()
        rSubject.update()
```

Verify that the relationship has been created

```python
for rSubject in mSubject.get_all():
    print(rSubject)
    for rEmg in rSubject.get_related():
        print(' '*PRETTY_INDENTATION,rEmg)
```

Now let's create relationships between the subject records and its files

```python
if DC_SUBJECT_TO_FILE_NAME in dcDataset.relationships().keys():
    fileRelationship = dcDataset.get_relationship(DC_SUBJECT_TO_FILE_NAME)
else:
    fileRelationship = dcDataset.create_relationship_type(DC_SUBJECT_TO_FILE_NAME,DC_SUBJECT_TO_FILE_DECRIPTION)
fileRelationship
```

For each subject, create relationship between subject record and subject files

```python
for sRecord in mSubject.get_all():
    # prepare the name of the collection of the subject
    sCollectionName = 'subject_' + str(sRecord.values['subject_id'])
    # get the collection
    sCollection = dcDataset.get_items_by_name(sCollectionName)[0]
    # loop on all the items in the collection and links them to the subject record
    for item in sCollection.items:
        # link the file to the record
        sRecord.relate_to(item,fileRelationship)
```

Verify that the relationship has been created

```python
mSubject.get(list(subjectVsEmg.keys())[0]).get_files()
for rSubject in mSubject.get_all():
    print(rSubject)
    for dp in rSubject.get_files():
        print(' '*PRETTY_INDENTATION,dp)
```
